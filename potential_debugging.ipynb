```python
import pandas as pd
import os
```

```python
folder_path = 'training-samples'
pathway_files = [f for f in os.listdir(folder_path) if f.endswith('.csv')]
print(f"Found {len(pathway_files)}")
```

```python
for file_name in pathway_files:
    file_path = os.path.join(folder_path, file_name)
    
    sample = pd.read_csv(file_path, sep='\t')
    print(file_name)
    print(sample[sample['label']==1])
    print('\n')
```
