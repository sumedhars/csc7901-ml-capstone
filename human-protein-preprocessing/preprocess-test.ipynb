```python
import pandas as pd
from Bio.KEGG.KGML.KGML_parser import read 
```

```python
pathway = read(open("hsa00010.xml", 'r'))
print(pathway)
```

```python
entries_data = []
for entry in pathway.entries.values():
    entries_data.append({
        'id': entry.id,
        'name': entry.name,
        'type': entry.type,
        'link': entry.link,
        'reaction': entry.reaction,
    })
entries_df = pd.DataFrame(entries_data)
entries_df
```

```python
human_genes_df = entries_df[entries_df['type'] == 'gene'] 
human_genes_df.head()
```

```python
for idx, row in human_genes_df.iterrows():
    gene_names = row['name'].split(' ')
    print(gene_names)
```
